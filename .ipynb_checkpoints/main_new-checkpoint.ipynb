# Run the pose detection model to extract the pose key point

```python
import util
img_in = cv2.imread(test_image_path)
img_in = cv2.resize(img_in, (500,500))
imageToTest_padded, pad = util.padRightDownCorner(img_in, 8, 128)        

img_in = np.transpose(np.float32(imageToTest_padded[:,:,:,np.newaxis]), (3,0,1,2))
output_blobs = model.predict(img_in) # 5x5x19 matrix heatmap
```

```python
pad
```

```python
heatmap = np.squeeze(output_blobs[1]) # output 1 is heatmaps
heatmap = cv2.resize(heatmap, (0,0), fx=8, fy=8, interpolation=cv2.INTER_CUBIC)
heatmap = cv2.resize(heatmap, (img_in.shape[1], img_in.shape[0]), interpolation=cv2.INTER_CUBIC)
print (heatmap.shape)

heatmap_avg = heatmap
all_peaks = []
peak_counter = 0
from scipy.ndimage.filters import gaussian_filter
for part in range(18):
    map_ori = heatmap_avg[:,:,part]
    map = gaussian_filter(map_ori, sigma=3)

    map_left = np.zeros(map.shape)
    map_left[1:,:] = map[:-1,:]
    map_right = np.zeros(map.shape)
    map_right[:-1,:] = map[1:,:]
    map_up = np.zeros(map.shape)
    map_up[:,1:] = map[:,:-1]
    map_down = np.zeros(map.shape)
    map_down[:,:-1] = map[:,1:]

    peaks_binary = np.logical_and.reduce((map>=map_left, map>=map_right, map>=map_up, map>=map_down, map > 0.1))
    peaks = list(zip(np.nonzero(peaks_binary)[1], np.nonzero(peaks_binary)[0])) # note reverse
    peaks_with_score = [x + (map_ori[x[1],x[0]],) for x in peaks]
    id = range(peak_counter, peak_counter + len(peaks))
    peaks_with_score_and_id = [peaks_with_score[i] + (id[i],) for i in range(len(id))]

    all_peaks.append(peaks_with_score_and_id)
    peak_counter += len(peaks)
```

```python
oriImg = cv2.imread(test_image_path) # B,G,R order
test_image = oriImg
multiplier = [1.0,1.0]
print(multiplier)
heatmap_avg = np.zeros((oriImg.shape[0], oriImg.shape[1], 19))


for m in range(len(multiplier)):
    scale = multiplier[m]
    imageToTest = cv2.resize(oriImg, (0,0), fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    imageToTest_padded = imageToTest

    input_img = np.transpose(np.float32(imageToTest_padded[:,:,:,np.newaxis]), (3,0,1,2)) # required shape (1, width, height, channels) 


    output_blobs = model.predict(input_img)

    # extract outputs, resize, and remove padding
    heatmap = np.squeeze(output_blobs[1]) # output 1 is heatmaps
    heatmap = cv2.resize(heatmap, (0,0), fx=8, fy=8, interpolation=cv2.INTER_CUBIC)
    heatmap = heatmap[:imageToTest_padded.shape[0]-pad[2], :imageToTest_padded.shape[1]-pad[3], :]
    heatmap = cv2.resize(heatmap, (oriImg.shape[1], oriImg.shape[0]), interpolation=cv2.INTER_CUBIC)
    print (heatmap.shape)

    heatmap_avg = heatmap_avg + heatmap / len(multiplier)

all_peaks = []
peak_counter = 0

for part in range(18):
    map_ori = heatmap_avg[:,:,part]
    map = gaussian_filter(map_ori, sigma=3)

    map_left = np.zeros(map.shape)
    map_left[1:,:] = map[:-1,:]
    map_right = np.zeros(map.shape)
    map_right[:-1,:] = map[1:,:]
    map_up = np.zeros(map.shape)
    map_up[:,1:] = map[:,:-1]
    map_down = np.zeros(map.shape)
    map_down[:,:-1] = map[:,1:]

    peaks_binary = np.logical_and.reduce((map>=map_left, map>=map_right, map>=map_up, map>=map_down, map > 0.1))
    peaks = list(zip(np.nonzero(peaks_binary)[1], np.nonzero(peaks_binary)[0])) # note reverse
    peaks_with_score = [x + (map_ori[x[1],x[0]],) for x in peaks]
    id = range(peak_counter, peak_counter + len(peaks))
    peaks_with_score_and_id = [peaks_with_score[i] + (id[i],) for i in range(len(id))]

    all_peaks.append(peaks_with_score_and_id)
    peak_counter += len(peaks)
```

# Visualize the key points

```python
# Draw the skeleton key points
colors = [[255, 0, 0], [255, 85, 0], [255, 170, 0], [255, 255, 0], [170, 255, 0], [85, 255, 0], [0, 255, 0], \
          [0, 255, 85], [0, 255, 170], [0, 255, 255], [0, 170, 255], [0, 85, 255], [0, 0, 255], [85, 0, 255], \
          [170, 0, 255], [255, 0, 255], [255, 0, 170], [255, 0, 85]]
ori_figure1 = cv2.imread(test_image_path)

for i in range(18):
    cv2.circle(ori_figure1, all_peaks[i][0][0:2], 4, colors[i], thickness=-1)

plt.imshow(ori_figure1[:,:,[2,1,0]])
```

```python
# Connect the skeleton key points in order
line_seg = [[0,1],[1,2],[2,3],[3,4],[1,5],[5,6],[6,7],[1,8],[8,9],[9,10],[1,11],[11,12],[12,13],[0,14],[0,15],[14,16],[15,17]]

ori_figure2 = cv2.imread(test_image_path) # B,G,R order
lineThickness = 2

for i in range(len(line_seg)):
    cv2.line(ori_figure2, all_peaks[line_seg[i][0]][0][0:2], all_peaks[line_seg[i][1]][0][0:2], (0,255,0), lineThickness) 

plt.imshow(ori_figure2[:,:,[2,1,0]])
```
